```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv("bhubaneswar_2018.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
plt.figure(figsize=[20,15])
sns.heatmap(data.isnull(),yticklabels=False,cmap='viridis')
```

```python
data.columns
```

```python
data.drop(['Unnamed: 0', 'wind_gust', 'wind_chill', 'heat_index', 'precip',
       'precip_rate', 'precip_total','cond', 'fog','snow', 'hail',
       'thunder', 'tornado','dir'],axis=1,inplace= True)
```

```python
data.head()
```

```python
data=data.set_index('date')
```

```python
data.index=pd.DatetimeIndex(data.index)
```

```python
data.head()
```

```python
data.index.date
```

```python
data.head(26)
```

```python
data_old=data.copy()
```

```python
plt.figure(figsize=[20,15])
sns.heatmap(data_old.isnull(),yticklabels=False,cmap='viridis')
```

```python
plt.figure(figsize=[20,10])
data_old.pressure.plot()
```

```python
plt.figure(figsize=[20,10])
data.pressure.fillna(data.pressure.rolling(104,min_periods=1).mean(),inplace=True)
data.pressure.plot()
```

```python
plt.figure(figsize=[20,10])
data_old.wind_spd.plot()
```

```python
plt.figure(figsize=[20,10])
data.wind_spd.fillna(data.wind_spd.rolling(26,min_periods=1).mean(),inplace=True)
data.wind_spd.plot()
```

```python
data.info()
```

```python
plt.figure(figsize=[20,10])
sns.heatmap(data.isnull(),yticklabels=False,cmap='viridis')
```

```python
data.to_csv('cleaned_Data.csv')
```
